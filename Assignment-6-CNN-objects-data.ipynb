```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Convolution2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Flatten
```

```python
from keras.preprocessing.image import ImageDataGenerator
train_datagen =  ImageDataGenerator(rescale = 1./255,shear_range = 0.2,zoom_range = 0.2,horizontal_flip = True)
test_datagen =  ImageDataGenerator(rescale = 1./255)
```

```python
x_train = train_datagen.flow_from_directory(r"C:\Users\Venu\Downloads\objects\things data\trainset",target_size = (64,64),batch_size = 32,class_mode = "categorical")
x_test = test_datagen.flow_from_directory(r"C:\Users\Venu\Downloads\objects\things data\testset",target_size = (64,64),batch_size = 32,class_mode = "categorical")
```

```python
model = Sequential()
```

```python
model.add(Convolution2D(32,(3,3) ,input_shape = (64,64,3),activation = "relu"))
```

```python
model.add(MaxPooling2D(2,2))
```

```python
model.add(Flatten())
```

```python
model.add(Dense(units = 512,activation = "relu",kernel_initializer = "random_uniform"))
```

```python
model.add(Dense(units = 5,activation = "softmax",kernel_initializer = "random_uniform"))
```

```python
model.compile("sgd",loss = "categorical_crossentropy",metrics = ["accuracy"])
```

```python
model.fit_generator(x_train ,steps_per_epoch = 3.125 ,epochs = 300,validation_data= x_test , validation_steps = 1.7)
```

```python
model.save("objects.h5")
```
